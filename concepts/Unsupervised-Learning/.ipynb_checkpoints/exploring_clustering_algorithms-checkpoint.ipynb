`Imports Libs`

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
```

```python
from sklearn.cluster import KMeans # KMeans Algorithms
from sklearn.cluster import AgglomerativeClustering # Agglomerative Algorithms
```

`Import Datasets`

```python
from sklearn import datasets
```

`Visualize Sample Datasets`

```python
# Import dataset with gaussian distribution
gaussian_features, gaussian_labels = datasets.make_gaussian_quantiles(n_samples=500, random_state=42)

# Import dataset with two-half moons
moons_features, moons_labels = datasets.make_moons(n_samples=150, random_state=42)

# Import dataset with circles 
circles_features, circles_labels = datasets.make_circles(n_samples=250, random_state=42)

# Importing dataset with clusters
cluster_features, cluster_labels = datasets.make_blobs(n_samples=250, n_features=3)
```

```python
# Function creates a scatterplot in two dimensions

def visualize_scatter_2D(data, title, hue=None):
    
    fig = plt.figure(figsize=(8, 5)) # Setting the figure size
    ax = sns.scatterplot(x=data[:,0], y=data[:, 1], hue=hue) # Scatterplot
    
    # Graph Attributes
    plt.title(title)
    plt.ylabel('Y')
    plt.xlabel('X')
```

```python
from mpl_toolkits.mplot3d import Axes3D

def visualize_scatter_3D(data, title, color=None):
    
    fig = plt.figure(figsize=(12, 8)) # Figure Size
    ax = plt.subplot(111, projection='3d') # Creating Axis
    
    ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=color)
    
    plt.title(title)
```

```python
# Gaussian Distribution Visual
visualize_scatter_2D(data=gaussian_features, title='Gaussian Distribution')
```

```python
# Two-Half Moons
visualize_scatter_2D(data=moons_features, title='Half Moons Distribution')
```

```python
# Circles Visual
visualize_scatter_2D(data=circles_features, title='Circles Distribution')
```

```python
# Clusters Visual
visualize_scatter_3D(data=cluster_features, title='Clustered Distribution')
```

`Fit Clustering Algorithms`

    a) KMeans - We will start by fitting the KMeans clustering algorithm on all 4 datasets.
    b) Agglomerative Clustering - Here we will fit the single link algorithm

`Instances of Clustering Algorithms`

```python
# KMeans
def KMeans_Algo(data):
    k_means = KMeans(n_clusters=2, random_state=10)
    cluster_preds = k_means.fit_predict(data)
    return cluster_preds

# Single-link
def single_link(data):
    single_link = AgglomerativeClustering(n_clusters=2, linkage='single')
    cluster_preds = single_link.fit_predict(data)
    return cluster_preds

# Complete-link
def complete_link(data):
    complete_link = AgglomerativeClustering(n_clusters=2, linkage='complete')
    cluster_preds = complete_link.fit_predict(data)
    return cluster_preds

# Average-link
def average_link(data):
    avg_link = AgglomerativeClustering(n_clusters=2, linkage='average')
    cluster_preds = avg_link.fit_predict(data)
    return cluster_preds

# Ward Method
def ward_method(data):
    ward = AgglomerativeClustering(n_clusters=2, linkage='ward')
    cluster_preds = ward.fit_predict(data)
    return cluster_preds
```

`Predictions on various datasets`

```python
# List of datasets
datasets = [gaussian_features, moons_features, circles_features, cluster_features]
dataset_names = ['Gaussian', 'Half-moons', 'Circles', 'Clustered']
```

```python
# KMeans Predictions
KMeans_preds = {}

index = 0
while index < len(datasets_features):
    preds = KMeans_Algo(data=datasets[index])
    KMeans_preds[datasets_name[index]] = preds
    
    index += 1
```

```python
# Single-link Predictions
Single_Link_preds = {}

index = 0
while index < len(datasets):
    preds = single_link(data=datasets[index])
    Single_Link_preds[datasets_name[index]] = preds
    
    index += 1
```

```python
# Average-link Predictions
Avg_Link_preds = {}

index = 0
while index < len(datasets):
    preds = average_link(data=datasets[index])
    Avg_Link_preds[datasets_name[index]] = preds
    
    index += 1
```

```python
# Complete-link Predictions
Complete_Link_preds = {}

index = 0
while index < len(datasets):
    preds = complete_link(data=datasets[index])
    Complete_Link_preds[datasets_name[index]] = preds
    
    index += 1
```

```python
# Single-link Predictions
ward_preds = {}

index = 0
while index < len(datasets):
    preds = ward_method(data=datasets[index])
    ward_preds[datasets_name[index]] = preds
    
    index += 1
```

`Study and compare algorithms`

`a) Gaussian Distribution`

```python
# List of gaussian predictions for the different algorithms
gaussian_preds = [KMeans_preds['Gaussian'], Single_Link_preds['Gaussian'], 
                  Avg_Link_preds['Gaussian'], Complete_Link_preds['Gaussian'], ward_preds['Gaussian']]

# Names of algorithms
algorithm_names = ['KMeans', 'Single-link', 'Average-link', 'Complete-link', 'Ward Method']

index = 0
while index < len(gaussian_preds):
    visualize_scatter_2D(
        data=gaussian_features, title='Gaussian Distribution - {}'.format(algorithm_names[index]), 
        hue=gaussian_preds[index])
    
    index += 1
```

```python
# List of gaussian predictions for the different algorithms
circles_preds = [KMeans_preds['Circles'], Single_Link_preds['Circles'], 
                  Avg_Link_preds['Circles'], Complete_Link_preds['Circles'], ward_preds['Circles']]

# Names of algorithms
algorithm_names = ['KMeans', 'Single-link', 'Average-link', 'Complete-link', 'Ward Method']

index = 0
while index < len(gaussian_preds):
    visualize_scatter_2D(
        data=circles_features, title='Circles Distribution - {}'.format(algorithm_names[index]), 
        hue=circles_preds[index])
    
    index += 1
```

```python
# List of gaussian predictions for the different algorithms
clustered_preds = [KMeans_preds['Clustered'], Single_Link_preds['Clustered'], 
                  Avg_Link_preds['Clustered'], Complete_Link_preds['Clustered'], ward_preds['Clustered']]

# Names of algorithms
algorithm_names = ['KMeans', 'Single-link', 'Average-link', 'Complete-link', 'Ward Method']

index = 0
while index < len(gaussian_preds):
    visualize_scatter_2D(
        data=cluster_features, title='Clustered Distribution - {}'.format(algorithm_names[index]), 
        hue=clustered_preds[index])
    
    index += 1
```
