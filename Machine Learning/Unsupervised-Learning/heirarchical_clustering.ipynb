#### Heirarchical Clustering

    Hierarchical clustering is a general family of clustering algorithms that build nested clusters by merging or splitting them successively. This hierarchy of clusters is represented as a tree (or dendrogram). The root of the tree is the unique cluster that gathers all the samples, the leaves being the clusters with only one sample. See the Wikipedia page for more details.

    The AgglomerativeClustering object performs a hierarchical clustering using a bottom up approach: each observation starts in its own cluster, and clusters are successively merged together. The linkage criteria determines the metric used for the merge strategy:

`Data Exploration`

```python
# "Iris" dataset
features, target = load_iris(return_X_y=True)
```

    The Iris dataset is classic dataset containing attributes of the Iris flower. In particular, the attributes
    are Sepal Length, Sepal Width, Petal Length and Petal Width. Furthermore, it contains a target for each
    flower's attributes that represents what family of Iris flower it belongs to.
    
    For the sake of clustering, we will disregard the target attribute and focus on the four attributes that 
    describe a flower.

```python
# Create DataFrame of data
iris_data = pd.DataFrame(features)

iris_data.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'] # Change column labels
iris_data.head()
```

```python
# Simple Stats
iris_data.describe()
```

```python
# Petal length vs Petal Width
plot = helper_functions.plot_scatter(data=iris_data, x=iris_data.Petal_Width, y=iris_data.Petal_Length)
```

```python
# Petal length vs Sepal length
plot = helper_functions.plot_scatter(data=iris_data, x=iris_data.Sepal_Length, y=iris_data.Petal_Length)
```

`Learning from the data`

```python
single_clustering = AgglomerativeClustering(n_clusters=3, linkage='single') # Instantiate Model
predictions = single_clustering.fit_predict(iris_data.values) # Fit and Predict cluster
```

```python
Z = hierarchy.linkage(np.reshape(predictions, (len(predictions), 1)), 'single')
dn = hierarchy.dendrogram(Z)

plt.figure()
```

        Finding the clusters with "single-link" produces an output of 3 clusters where one of the clusters
        is significantly small.
        
        Based on our knowledge of the targets, this is not the best approach to clustering the data.
